```python
def convert_units(conversion_type, value):
    """
    Converts units based on the selected conversion type.

    Parameters:
    conversion_type (str): The type of conversion selected by the user.
    value (float): The numeric value that needs to be converted.

    Returns:
    float: The converted value.
    """

    if conversion_type == "1":
        return value * 3.28084
    elif conversion_type == "2":
        return value / 3.28084
    elif conversion_type == "3":
        return value * 2.20462
    elif conversion_type == "4":
        return value / 2.20462
    elif conversion_type == "5":
        return value * 0.264172
    elif conversion_type == "6":
        return value / 0.264172
    else:
        return None



print("Unit Conversion Program")
print("------------------------")
print("1. Meters (m) to Feet (ft)")
print("2. Feet (ft) to Meters (m)")
print("3. Kilograms (kg) to Pounds (lbs)")
print("4. Pounds (lbs) to Kilograms (kg)")
print("5. Liters (L) to Gallons (gal)")
print("6. Gallons (gal) to Liters (L)")

conversion_type = input("Choose a conversion option (1-6): ")

try:
    value = float(input("Enter the value to convert: "))

    result = convert_units(conversion_type, value)

    if result is None:
        print("Error: Unsupported conversion type.")
    else:
        print(f"Converted value: {result:.2f}")

except ValueError:
    print("Error: Please enter a valid numeric value.")
```

```python
def find_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The sum of all numbers in the list.
    """
    return sum(numbers)


def find_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The average of the numbers in the list.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The largest number in the list.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The smallest number in the list.
    """
    return min(numbers)


print("Mathematical Operations on a List")
print("---------------------------------")
print("1. Find Sum")
print("2. Find Average")
print("3. Find Maximum")
print("4. Find Minimum")

choice = input("Choose an operation (1-4): ")

try:
    user_input = input("Enter numbers separated by spaces: ")
    numbers = [float(num) for num in user_input.split()]

    if len(numbers) == 0:
        raise ValueError("Empty list")

    if choice == "1":
        result = find_sum(numbers)
        print("Sum:", result)
    elif choice == "2":
        result = find_average(numbers)
        print("Average:", result)
    elif choice == "3":
        result = find_maximum(numbers)
        print("Maximum:", result)
    elif choice == "4":
        result = find_minimum(numbers)
        print("Minimum:", result)
    else:
        print("Error: Invalid operation selected.")

except ValueError:
    print("Error: Please enter valid numbers separated by spaces.")
```

## 1. Extract Every Other Element

```python
def extract_every_other(lst):
    """
    Returns every other element from the given list, starting from the first element.

    Parameters:
    lst (list): The input list from which elements will be extracted.

    Returns:
    list: A new list containing every other element from the original list.
    """
    return lst[::2] # Slicing the list by setting the step to 2


numbers = [1, 2, 3, 4, 5, 6]
result = extract_every_other(numbers)

print("Original list:", numbers)
print("Every other element:", result)
```

## 2. Slice a Sublist:

```python
def get_sublist(lst, start, end):
    """
    Returns a sublist from the given list starting at index 'start'
    and ending at index 'end' (inclusive).

    Parameters:
    lst (list): The original list.
    start (int): The starting index.
    end (int): The ending index.

    Returns:
    list: A sublist from start to end.
    """
    return lst[start:end+1]


numbers = [1, 2, 3, 4, 5, 6]
result = get_sublist(numbers, 2, 4)

print("Original list:", numbers)
print("Sublist:", result)
```

## 3. Reverse a List Using Slicing:

```python
def reverse_list(lst):
    """
    Returns the reversed version of the given list using slicing.

    Parameters:
    lst (list): The list to be reversed.

    Returns:
    list: A new list containing the elements of the original list in reverse order.
    """
    return lst[::-1]


numbers = [1, 2, 3, 4, 5]
result = reverse_list(numbers)

print("Original list:", numbers)
print("Reversed list:", result)
```

## 4. Remove the First and Last Elements:

```python
def remove_first_last(lst):
    """
    Removes the first and last elements from a list.

    Parameters:
    lst (list): The original list.

    Returns:
    list: A new list without the first and last elements.
    """
    return lst[1:-1]


numbers = [1, 2, 3, 4, 5]
result = remove_first_last(numbers)

print("Original list:", numbers)
print("After removing first and last:", result)
```

## 5. Get the First n Elements:

```python
def get_first_n(lst, n):
    """
    Returns the first n elements from a list.

    Parameters:
    lst (list): The original list.
    n (int): The number of elements to extract.

    Returns:
    list: A list containing the first n elements.
    """
    return lst[:n]


numbers = [1, 2, 3, 4, 5]
result = get_first_n(numbers, 3)

print("Original list:", numbers)
print("First n elements:", result)
```

## 6. Extract Elements from the End:

```python
def get_last_n(lst, n):
    """
    Returns the last n elements from a list.

    Parameters:
    lst (list): The original list.
    n (int): The number of elements to extract from the end.

    Returns:
    list: A list containing the last n elements.
    """
    return lst[-n:]


numbers = [1, 2, 3, 4, 5]
result = get_last_n(numbers, 2)

print("Original list:", numbers)
print("Last n elements:", result)
```

## 7. Extract Elements in Reverse Order:

```python
def reverse_skip(lst):
    """
    Returns every second element from the list starting from the second-to-last element,
    moving backward.

    Parameters:
    lst (list): The original list.

    Returns:
    list: A new list containing elements in reverse order, skipping one element each time.
    """
    return lst[-2::-2]


numbers = [1, 2, 3, 4, 5, 6]
result = reverse_skip(numbers)

print("Original list:", numbers)
print("Reversed skip list:", result)
```

# 4.3 Exercise on Nested List:

## 1. Flatten a Nested List:

```python
def flatten(lst):
    """
    Flattens a nested list into a single list with all elements in one dimension.

    Parameters:
    lst (list): A nested list (list of lists).

    Returns:
    list: A flattened list containing all elements from the nested lists.
    """
    flat_list = []
    for sublist in lst:
        for item in sublist:
            flat_list.append(item)
    return flat_list


nested_list = [[1, 2], [3, 4], [5]]
result = flatten(nested_list)

print("Original nested list:", nested_list)
print("Flattened list:", result)
```

## 2. Accessing Nested List Elements:


```python
def access_nested_element(lst, indices):
    """
    Accesses an element from a nested list using a list of indices.

    Parameters:
    lst (list): The nested list.
    indices (list): A list of indices to access the desired element.

    Returns:
    element: The element at the specified position in the nested list.
    """
    element = lst
    for index in indices:
        element = element[index]
    return element


nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
result = access_nested_element(nested_list, indices)

print("Nested list:", nested_list)
print("Indices:", indices)
print("Accessed element:", result)
```

## 3. Sum of All Elements in a Nested List:

```python
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list, including nested sublists.

    Parameters:
    lst (list): A nested list containing numbers and/or other lists.

    Returns:
    int/float: The sum of all numbers in the nested list.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

nested_list = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(nested_list)

print("Nested list:", nested_list)
print("Sum of all elements:", result)
```

## 4. Remove Specific Element from a Nested List:

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
    lst (list): The nested list.
    elem: The element to remove.

    Returns:
    list: The modified list with all occurrences of elem removed.
    """
    new_list = []
    for item in lst:
        if isinstance(item, list):
            new_list.append(remove_element(item, elem))
        else:
            if item != elem:
                new_list.append(item)
    return new_list


nested_list = [[1, 2], [3, 2], [4, 5]]
result = remove_element(nested_list, 2)

print("Original list:", nested_list)
print("Modified list:", result)
```

## 5. Find the Maximum Element in a Nested List:

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list, including all sublists.

    Parameters:
    lst (list): A nested list containing numbers and/or other lists.

    Returns:
    int/float: The maximum value in the nested list.
    """
    max_val = 0

    for item in lst:
        if isinstance(item, list):
            sub_max = find_max(item)
            if sub_max > max_val:
                max_val = sub_max
        else:
            if item > max_val:
                max_val = item

    return max_val

nested_list = [[1, 2], [3, [4, 5]], 6]
result = find_max(nested_list)

print("Nested list:", nested_list)
print("Maximum element:", result)
```

## 6. Count Occurrences of an Element in a Nested List:


```python
def count_occurrences(lst, elem):
    """
    Counts how many times a specific element appears in a nested list.

    Parameters:
    lst (list): A nested list containing numbers and/or other lists.
    elem: The element to count.

    Returns:
    int: The number of occurrences of elem in the nested list.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count

nested_list = [[1, 2], [2, 3], [2, 4]]
element = 2
result = count_occurrences(nested_list, element)

print("Nested list:", nested_list)
print(f"Occurrences of {element}:", result)
```

## 7. Flatten a List of Lists of Lists:

```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list, regardless of depth.

    Parameters:
    lst (list): A nested list (can contain lists within lists at any depth).

    Returns:
    list: A flattened list containing all elements.
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))
        else:
            flat_list.append(item)
    return flat_list

nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(nested_list)

print("Original nested list:", nested_list)
print("Flattened list:", result)
```

## 8. Nested List Average:

```python
def flatten_and_sum(lst):
    """Helper function: returns (total_sum, total_count) of elements in nested list"""
    total_sum = 0
    total_count = 0
    for item in lst:
        if isinstance(item, list):
            sub_sum, sub_count = flatten_and_sum(item)
            total_sum += sub_sum
            total_count += sub_count
        else:
            total_sum += item
            total_count += 1
    return total_sum, total_count

def average_nested(lst):
    """
    Calculates the average of all elements in a nested list, regardless of depth.

    Parameters:
    lst (list): A nested list containing numbers and/or other lists.

    Returns:
    float: The average of all numbers in the nested list.
    """

    total_sum, total_count = flatten_and_sum(lst)
    if total_count == 0:
        return 0
    return total_sum / total_count

nested_list = [[1, 2], [3, 4], [5, 6]]
result = average_nested(nested_list)

print("Nested list:", nested_list)
print("Average of all elements:", result)
```

## Matrix Multiplication

```python
n = 1000

A = [[1]*n for _ in range(n)]
B = [[1]*n for _ in range(n)]
start = time.time()
result = [[0]*n for _ in range(n)]
for i in range(n):
    for j in range(n):
        for k in range(n):
            result[i][j] += A[i][k] * B[k][j]
end = time.time()
print("Python list matrix multiplication time:", end - start)
```
